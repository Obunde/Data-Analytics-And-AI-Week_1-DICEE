#Calculating Efficiency of Fuel Depot

## Calculating Fuel Depot Efficiency

`calculate_efficiency` measures how well a depot performed against its target:

    efficiency = (actual_output / target_output) * 100

Result is rounded to 2 decimal places for clean output, and returns 0 if `target_output` is 0 (avoids a crash from dividing by zero).

We'll reuse this function to loop through all depots next.

```python
#defining KPI function
def calculate_efficiency(actual_output,target_output):
    if target_output == 0:
        return 0
    efficiency = actual_output/target_output * 100
    return round( efficiency,2)
```

## Classifying Operational Status

`get_operational_status` takes an efficiency percentage and labels it into one of three bands:

- **< 70** → `"critical"` — depot is significantly under target, needs urgent attention
- **70–89** → `"warning"` — depot is below target, keep an eye on it
- **≥ 90** → `"normal"` — depot is performing as expected

This gives us a quick, human-readable label instead of having to compare raw numbers every time.

```python
def get_operational_status(efficiency):
    if efficiency < 70:
        return "critical"
    elif 70 <= efficiency <90:
        return "warning"
    else:
        return "normal"
```

## Site Data

`depots` stores raw output data for each depot. Each key is a depot name, and each value is a dictionary holding:

- `actual_output`: units actually produced
- `target_output`: units the depot was expected to produce

This is the raw input we'll feed into `calculate_efficiency` and `get_operational_status` to evaluate each depot's performance.

```python
# Site Data
depots = {
"Nairobi Depot":{
    "actual_output":800,
    "target_output":1000
    },
"Nakuru Depot":{
    "actual_output":400,
    "target_output":1000
    },
"Mombasa Depot":{
    "actual_output":950,
    "target_output":1000
    }
}
```

```python
## Evaluating Each Depot

Now we bring everything together. For every depot in our `depots` dictionary, we:

1. Pull out `actual_output` and `target_output`
2. Calculate its `efficiency` using `calculate_efficiency`
3. Classify that efficiency into a `status` using `get_operational_status`
4. Print a summary line for that depot

This loop is where the individual functions we built earlier actually get used together to produce a full performance report.
```

```python
# Looping through each depot to calculate efficiency and status
for depot, data in depots.items():
    actual = data["actual_output"]
    target = data["target_output"]
    efficiency = calculate_efficiency(actual, target)
    status = get_operational_status(efficiency)

    # Pick an emoji based on status, for quick visual scanning
    if status == "critical":
        status_emoji = "🔴"
    elif status == "warning":
        status_emoji = "🟡"
    else:
        status_emoji = "🟢"

    print(f"Depot Name : {depot}")
    print(f"Actual Output  = {actual}")
    print(f"Target Output  = {target}")
    print(f"Efficiency     = {efficiency}%")
    print(f"  {status_emoji} Status         : {status}")
    print()  # blank line between depots
```
